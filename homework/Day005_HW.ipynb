```python

import requests
r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response=r.text
response

```

```python
#這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import json 
data= json.loads(response)
data1=data[0]
print('回傳',len(data),'筆資料')
a=[]
for d in data1.keys():
    a.append(d)
print('每一筆資料包含:',a)
```

```python
#取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for d in data:
    print(d['title'],d['createdAt'],'按讚:',d['commentCount'],'留言:',d['likeCount'])
```

```python
#計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
import requests
r2=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response2=r2.text
response2
```

```python
data2= json.loads(response2)
data2
print(len(data2))
```

```python
li=[]
li2=[]
for d in data:
    li.append(d['likeCount'])
    li2.append(d['commentCount'])
```

```python
print('熱門文章:')
print('平均按讚人數:',float(sum(li)/31))
print('平均留言人數:',float(sum(li2)/31))
```

```python
li3=[]
li4=[]
for d in data2:
    li3.append(d['likeCount'])
    li4.append(d['commentCount'])
```

```python
print('非熱門文章:')
print('平均按讚人數:',float(sum(li3)/31))
print('平均留言人數:',float(sum(li4)/31))
```
