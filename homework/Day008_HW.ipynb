```python
#Request 取回之後該怎麼取出資料，資料型態是什麼？
print('用.text',',','資料型態為字串')
```

```python
import requests # 引入函式庫
r = requests.get('https://github.com/timeline.json') # 想要爬資料的目標網址
response = r.text # 模擬發送請求的動作
print(type(response))
```

```python
#為什麼要使用 BeatifulSoup 處理？處理後的型態是什麼？
print('使得http依照各個標籤排列，便於資料內容的閱讀')
print('資料型態為bs4.BeautifulSoup')
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(response,"html5lib")
print(soup)
```

```python
#觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？
rr=requests.get('https://www.zhihu.com/explore')
response2=rr.text
soup=BeautifulSoup(response2,"html5lib")
soup
print('加headers')
```

```python
headers = {'user-agent': 'my-app/0.0.1'}
rr=requests.get('https://www.zhihu.com/explore',headers=headers)
response2=rr.text
soup=BeautifulSoup(response2,"html5lib")
soup
```
